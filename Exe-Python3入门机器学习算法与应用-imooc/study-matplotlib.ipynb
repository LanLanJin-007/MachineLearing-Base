```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funy(x):
    if(abs(x) <= 1):
        y = 2-x**2
    elif(abs(x)<= 2):
        y = (abs(x)-2)**2
    else:
        y = 0
    return y
```

```python
X = np.linspace(-3,3,2000)
Y = [funy(x) for x in X ]
plt.plot(X,Y,'r')
plt.show()
```
